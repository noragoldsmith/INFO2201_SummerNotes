```python
import csv
```

```python
myDict = {}

with open("data/2024-annual-report-report-data-all-states.csv", "r", encoding = "utf-8", errors = "replace") as file:

    csvReader = csv.reader(file, delimiter = ",")
    next(csvReader)

    for row in csvReader:
        state = row[2]
        factor = row[1]
        
        if state not in myDict:
            myDict[state] = {}

    
        if factor == "Depression":
            if factor not in myDict[state]:
                myDict[state][factor] = {}
                myDict[state][factor]["Rank"] = row[5]
                myDict[state][factor]["Value"] = row[3]
    
    
        if factor == "Mental Illness in Household":
            if factor not in myDict[state]:
                myDict[state][factor] = {}
                myDict[state][factor]["Rank"] = row[5]
                myDict[state][factor]["Value"] = row[3]
    
            
    
```
