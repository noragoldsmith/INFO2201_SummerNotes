## CSV Files

### Exercise 3: 

import tv_shows.csv and find the 5 most recently made tv shows

```python
import csv

showList = []
with open("data/tv_shows.csv", "r") as f:
    csvReader = csv.reader(f, delimiter = ",", quotechar = '"')
    next(csvReader)

    for row in csvReader:
        showList.append(row)

showList
```

```python
yrList = []
for show in showList:
    yrList.append(show[2].replace("(", "").replace(")", ""))
```

## Exercise 4 
import nationalParks.json 

and make a dictionary that has the index in the json file an the name of the national park

```python
myDict = {}

f = open("data/nationalParks.json", "r")
data = json.loads(f.read())
    
```

```python
pprint.pprint(data)
```

```python
for item in data:
    key = data.index(item)
    value = item["title"]

    myDict[key] = value
```

```python
myDict
```

## Exercise 5

get the name (title) of the park with the most visitors

Hint: You will need to convert the numbers to int and use **replace** to deal with commas

```python
newDict = {}

for item in data:
    parkName = item["title"]
    numVisitors = item["visitors"]
    numVisitorsInt = int(numVisitors.replace(",", ""))
    newDict[parkName] = numVisitorsInt

newDict
```

```python
maxVisitors = max(newDict.values())

for parkName, numVisitorsInt in newDict.items():
    if numVisitorsInt == maxVisitors:
        print(parkName)
```

## Exercise 6
Write a function that recieves the name of the park and prints out it's state, establishment date, and description

```python
nameOfPark = input("Name of park: ")

for park in data:
    if nameOfPark == park["title"]:
        print(park["states"][0]["title"])
        print(park["date_established_readable"])
        print(park["description"])
```
